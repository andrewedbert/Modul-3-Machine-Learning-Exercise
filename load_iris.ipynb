```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
dataIris = load_iris()
# dataIris
# dir(dataIris)
# dataIris['feature_names']
dataIris['target_names']
```

```python
dfIris = pd.DataFrame(
    dataIris['data'],
    columns = dataIris['feature_names']
)
dfIris['target'] = dataIris['target']
dfIris['species'] = dfIris['target'].apply(
    lambda x: dataIris['target_names'][x]
)
dfIris
```

```python
dfSetosa = dfIris[dfIris['target']==0]
dfVersicolor = dfIris[dfIris['target']==1]
dfVirginica = dfIris[dfIris['target']==2]
```

```python
fig = plt.figure('Data Iris', figsize = (12,6))

# plot sepal length vs sepal width
plt.subplot(121)
plt.scatter(dfSetosa['sepal length (cm)'], dfSetosa['sepal width (cm)'], marker='o', color='r')
plt.scatter(dfVersicolor['sepal length (cm)'], dfVersicolor['sepal width (cm)'], marker='o', color='g')
plt.scatter(dfVirginica['sepal length (cm)'], dfVirginica['sepal width (cm)'], marker='o', color='b')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal width (cm)')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.grid(True)

# plot petal length vs petal width
plt.subplot(122)
plt.scatter(dfSetosa['petal length (cm)'], dfSetosa['petal width (cm)'], marker='o', color='r')
plt.scatter(dfVersicolor['petal length (cm)'], dfVersicolor['petal width (cm)'], marker='o', color='g')
plt.scatter(dfVirginica['petal length (cm)'], dfVirginica['petal width (cm)'], marker='o', color='b')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.grid(True)

plt.show()
```
