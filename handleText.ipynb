# Dummy Variables

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

#### 1. read dataset (data.csv)

```python
data = pd.read_csv('data.csv')
data
```

#### 2. create dummy variable for data['kota']

```python
dataDum = pd.get_dummies(data['kota'])
dataDum
```

#### 3. create final dataframe

```python
dataFinal = pd.concat([data, dataDum], axis = 1)
# dataFinal.drop(['kota'], axis = 1, inplace = True)
dataFinal = dataFinal.drop(['kota'], axis = 1)
dataFinal
```

#### 4. linear regression

```python
y = dataFinal['harga']
x = dataFinal.drop(['harga'], axis = 'columns')
x
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
model.score(x, y) * 100
```

```python
# luas 1000 di Depok
model.predict([[1000, 0,1,0]])[0]
```

```python
# Luas 1000 di Bogor
model.predict([[1000, 1,0,0]])[0]
```

```python
# Luas 1000 di Jakarta
model.predict([[1000, 0,0,1]])[0]
```
