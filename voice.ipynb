```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

#### 1. Read _voice.csv_

Kaggle voice dataset from [here](https://www.kaggle.com/primaryobjects/voicegender)

```python
df = pd.read_csv('voice.csv')
# df.isnull().sum()
df['label2'] = df['label'].apply(
    lambda i: 1 if i == 'male' else 0
)
corr = df.corr()
df.shape
```

```python
from sklearn.model_selection import train_test_split
xtr, xts, ytr, yts = train_test_split(
    df.drop('label', axis = 1),
    df['label'],
    test_size = .1,
#     random_state = 1
)
xtr.head()
```

```python
model = LogisticRegression(
    solver='lbfgs',
    multi_class='auto',
    max_iter=1000
)
model.fit(xtr,ytr)
x = model.predict(df.drop('label', axis = 1))
df['prediksi'] = x
df
```
