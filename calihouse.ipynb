```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataCH = fetch_california_housing()
# dir(dataCH)
# dataCH.data.shape
# dataCH['feature_names']
dataCH.target[0]
```

```python
dfCH = pd.DataFrame(
    dataCH.data,
    columns = dataCH['feature_names']
)
dfCH['Price'] = dataCH.target
dfCH
```

```python
plt.imshow(dfCH.corr())
plt.yticks(np.arange(9), dfCH.columns)
plt.colorbar()
```

```python
dfCH.corr()
```

## Splitting datasets: 
#### 90% training = 10% testing

```python
x = dfCH[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude']]
y = dfCH['Price']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.1)

# dfCH.columns
x_test.iloc[0]
```

```python
y_test.iloc[0]
```

## Linear Regression

```python
model = LinearRegression()
model.fit(x_train, y_train)

# prediksi
# model.predict([x_test.iloc[0]])

# score accuracy
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
